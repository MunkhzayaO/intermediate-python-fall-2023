```python
class Plot:
    def __init__(self, dataframe):
       
        self.dataframe = dataframe

        # Creating a box plot
    def box_plots(self):
        
        numeric_features = self.dataframe.select_dtypes(include=['number'])
        num_features_count = len(numeric_features.columns)

        fig, axes = plt.subplots(nrows=1, ncols=num_features_count, figsize=(15, 5))
        fig.suptitle("Box Plots of Numeric Features")

        for i, col in enumerate(numeric_features.columns):
            sns.boxplot(x=self.dataframe[col], ax=axes[i])
            axes[i].set_title(col)

        plt.show()

        # Creating a scatter plot
    def scatter_plot(self, column1, column2):
    
        plt.scatter(self.dataframe[column1], self.dataframe[column2])
        plt.xlabel(column1)
        plt.ylabel(column2)
        plt.title(f"Scatter Plot: {column1} vs {column2}")
        plt.show()
        
        
        # Creating a scatter matrix
    def scatter_matrix_with_histograms(self):
        
        # Selecting only numeric columns
        numeric_features = self.dataframe.select_dtypes(include=['number'])

      
        sns.pairplot(numeric_features, diag_kind="hist")

        plt.show()

    def plot(self):
        
        self.box_plots()
        self.scatter_plot(self.dataframe.columns[0], self.dataframe.columns[1])
        self.scatter_matrix_with_histograms()


if __name__ == "__main__":
    # Loading the dataset
    df = pd.read_csv('./data/titanic.csv')

    plotter = Plot(df)

    # 1
    plotter.box_plots()

    # 2
    plotter.scatter_plot('Age', 'Fare')

    # 3
    plotter.scatter_matrix_with_histograms()
```
